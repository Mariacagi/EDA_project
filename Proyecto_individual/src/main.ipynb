```python
# Copia para trabajar con ella
df_amazon_copy = df_amazon.copy()
```

```python
# Se pasa la columna price a string
df_amazon_copy["price"].astype(str)
```

```python
# Se separan los precios, el más bajo y el más alto
    # Ojo!! No todas las columnas tienen un segundo precio (el precio alto)

price_str = split_columna(df = df_amazon_copy, columna = "price", objeto_separacion = " ", numero_veces=1)
df_amazon_copy["low_price"]= price_str[0]
df_amazon_copy["high_price"]=price_str[1]
df_amazon_copy
```

```python
# Se quita el $ y se hace float
df_amazon_copy['low_price'] = df_amazon_copy['low_price'].str.replace('$', '').astype(float)
df_amazon_copy['high_price'] = df_amazon_copy['high_price'].str.replace('$', '').astype(float)
df_amazon_copy.sort_values("low_price")
```

```python
# Se reemplaza por cero los valores NaN en la columna low_price
df_amazon_copy['low_price'] = df_amazon_copy['low_price'].replace(np.nan, 0)
df_amazon_copy.sort_values('low_price')
```

```python
# Se ordena por low_price y se eliminan duplicados cogiendo el precio más alto
    # Se coge low_price ya que high_price tiene NaN, porque no todos los items de la columna price tenían doble precio
new_df_amazon = df_amazon_copy.sort_values('low_price', ascending=False).drop_duplicates('ASIN', keep="first").sort_index()
new_df_amazon = new_df_amazon.reset_index(drop=True)
new_df_amazon.sort_values("low_price")
```

## Comprobación de valores NaN de nuevo

```python
# Se pasa la columna price a str para poder iterar sobre ella
new_df_amazon["price"] = new_df_amazon["price"].astype(str)
```

```python
# Se itera sobre price. Si no hay un precio en la columna(es decir = nan), reemplazamos 0 por NaN para poder contar nulls
for pos, x in enumerate(new_df_amazon.price):
    if "$" not in x:
        new_df_amazon['low_price'][pos] = np.nan

new_df_amazon.sort_values("low_price")
```

```python
# Por tanto ahora, sin duplicados y pasando de nuevo los 0 a null tenemos menos valores null
total_nan_values = new_df_amazon.low_price.isnull().sum().sum()
print (total_nan_values)
```
